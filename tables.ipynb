```python
import pandas as pd
import numpy as np
from statistics import mean
```

```python
#import results from all alg/data set combos

#results from Letter 1
lr_letter1 = pd.read_csv("lr_letter1.csv").set_index("Unnamed: 0")
rf_letter1 = pd.read_csv("rf_letter1.csv").set_index("Unnamed: 0")
knn_letter1 = pd.read_csv("knn_letter1.csv").set_index("Unnamed: 0")

#results from Letter 2
lr_letter2 = pd.read_csv("lr_letter2.csv").set_index("Unnamed: 0")
rf_letter2 = pd.read_csv("rf_letter2.csv").set_index("Unnamed: 0")
knn_letter2 = pd.read_csv("knn_letter2.csv").set_index("Unnamed: 0")

#results from Cover_Type
lr_cov = pd.read_csv("lr_cov.csv").set_index("Unnamed: 0")
rf_cov = pd.read_csv("rf_cov.csv").set_index("Unnamed: 0")
knn_cov = pd.read_csv("knn_cov.csv").set_index("Unnamed: 0")

#results from Mushroom
lr_mush = pd.read_csv("lr_mush.csv").set_index("Unnamed: 0")
rf_mush = pd.read_csv("rf_mush.csv").set_index("Unnamed: 0")
knn_mush = pd.read_csv("knn_mush.csv").set_index("Unnamed: 0")

display(lr_letter1, rf_letter1, knn_letter1)
```

```python
# Appendix Table 2 raw test scores
lr_raw_test_scores = lr_letter1[['Test ACC', 'Test F1', 'Test AUC']]

raw_test_scores = pd.concat([lr_letter1[['Test ACC', 'Test F1', 'Test AUC']],
                             rf_letter1[['Test ACC', 'Test F1', 'Test AUC']],
                             knn_letter1[['Test ACC', 'Test F1', 'Test AUC']],
                             lr_letter2[['Test ACC', 'Test F1', 'Test AUC']],
                             rf_letter2[['Test ACC', 'Test F1', 'Test AUC']],
                             knn_letter2[['Test ACC', 'Test F1', 'Test AUC']],
                             lr_cov[['Test ACC', 'Test F1', 'Test AUC']],
                             rf_cov[['Test ACC', 'Test F1', 'Test AUC']],
                             knn_cov[['Test ACC', 'Test F1', 'Test AUC']],
                             lr_mush[['Test ACC', 'Test F1', 'Test AUC']],
                             rf_mush[['Test ACC', 'Test F1', 'Test AUC']],
                             knn_mush[['Test ACC', 'Test F1', 'Test AUC']]
                            ],
                            axis=0)
raw_test_scores.index.names = ['Algo-Dataset-Trial']
raw_test_scores.to_latex(label = 'Raw test set performance')

raw_test_scores.to_latex() #for writeup
```

```python
# Appendix Table 1 mean training scores
lr_letter1_mean = pd.DataFrame(lr_letter1.mean().to_dict(),index=[lr_letter1.index.values[-1]])
rf_letter1_mean = pd.DataFrame(rf_letter1.mean().to_dict(),index=[rf_letter1.index.values[-1]])
knn_letter1_mean = pd.DataFrame(knn_letter1.mean().to_dict(),index=[knn_letter1.index.values[-1]])

lr_letter2_mean = pd.DataFrame(lr_letter2.mean().to_dict(),index=[lr_letter2.index.values[-1]])
rf_letter2_mean = pd.DataFrame(rf_letter2.mean().to_dict(),index=[rf_letter2.index.values[-1]])
knn_letter2_mean = pd.DataFrame(knn_letter2.mean().to_dict(),index=[knn_letter2.index.values[-1]])

lr_cov_mean = pd.DataFrame(lr_cov.mean().to_dict(),index=[lr_cov.index.values[-1]])
rf_cov_mean = pd.DataFrame(rf_cov.mean().to_dict(),index=[rf_cov.index.values[-1]])
knn_cov_mean = pd.DataFrame(knn_cov.mean().to_dict(),index=[knn_cov.index.values[-1]])

lr_mush_mean = pd.DataFrame(lr_mush.mean().to_dict(),index=[lr_mush.index.values[-1]])
rf_mush_mean = pd.DataFrame(rf_mush.mean().to_dict(),index=[rf_mush.index.values[-1]])
knn_mush_mean = pd.DataFrame(knn_mush.mean().to_dict(),index=[knn_mush.index.values[-1]])


mean_scores = pd.concat([lr_letter1_mean, rf_letter1_mean, knn_letter1_mean, 
                               lr_letter2_mean, rf_letter2_mean, knn_letter2_mean,
                               lr_cov_mean, rf_cov_mean, knn_cov_mean,
                               lr_mush_mean, rf_mush_mean, knn_mush_mean], axis = 0)


mean_scores[['Train ACC', 'Train F1', 'Train AUC']].to_latex()
```

```python
# mean test scores:
mean_test = mean_scores[['Test ACC', 'Test F1', 'Test AUC']]
mean_test
```

```python
# computing averages for main table 1
# each line gets the value for a single cell in the table

#cols of row 1
lr_avg_acc = (mean_test['Test ACC'][0] +mean_test['Test ACC'][3] 
              +mean_test['Test ACC'][6] +mean_test['Test ACC'][9])/4
        
lr_avg_f1 = (mean_test['Test F1'][0] + mean_test['Test F1'][3] 
              +mean_test['Test F1'][6] +mean_test['Test F1'][9])/4
        
lr_avg_auc = (mean_test['Test AUC'][0] + mean_test['Test AUC'][3] 
              +mean_test['Test AUC'][6] +mean_test['Test AUC'][9])/4


#cols of row 2
rf_avg_acc = (mean_test['Test ACC'][1] +mean_test['Test ACC'][7] 
              +mean_test['Test ACC'][4] +mean_test['Test ACC'][10])/4
        
rf_avg_f1 = (mean_test['Test F1'][1] + mean_test['Test F1'][7] 
              +mean_test['Test F1'][4] +mean_test['Test F1'][10])/4
        
rf_avg_auc = (mean_test['Test AUC'][1] + mean_test['Test AUC'][7] 
              +mean_test['Test AUC'][4] +mean_test['Test AUC'][10])/4

#cols of row 3
knn_avg_acc = (mean_test['Test ACC'][2] +mean_test['Test ACC'][5] 
              +mean_test['Test ACC'][8] +mean_test['Test ACC'][11])/4
        
knn_avg_f1 = (mean_test['Test F1'][2] + mean_test['Test F1'][5] 
              +mean_test['Test F1'][8] +mean_test['Test F1'][11])/4
        
knn_avg_auc = (mean_test['Test AUC'][2] + mean_test['Test AUC'][5] 
              +mean_test['Test AUC'][8] +mean_test['Test AUC'][11])/4

data = {'ACC':[lr_avg_acc,rf_avg_acc,knn_avg_acc ], 
        'F1':[lr_avg_f1, rf_avg_f1, knn_avg_f1],
        'AUC':[lr_avg_auc,rf_avg_auc, knn_avg_auc]} 
table_1 = pd.DataFrame(data, index =['LR', 'RF', 'KNN']) 
display(table_1)
table_1.to_latex() #get latex code
```

```python
# computing averages for main table 2
# each line accesses the test, acc, auc scores for a given algo/data set combo
# take the mean across those metrics to get a single value for a cell in a table

# col 1 = letter 1
lr_avg_lttr1 = mean(lr_letter1_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0]) 
rf_avg_lttr1 = mean(rf_letter1_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
knn_ag_lttr1 = mean(knn_letter1_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])

# col 2 = letter 2
lr_avg_lttr2 = mean(lr_letter2_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
rf_avg_lttr2 = mean(rf_letter2_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
knn_ag_lttr2 = mean(knn_letter2_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])

# col 3 = cov_type
lr_avg_cov = mean(lr_cov_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
rf_avg_cov = mean(rf_cov_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
knn_ag_cov = mean(knn_cov_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])

# col 4 = mushroom
lr_avg_mush = mean(lr_mush_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
rf_avg_mush = mean(rf_mush_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])
knn_ag_mush = mean(knn_mush_mean[['Test ACC', 'Test F1', 'Test AUC']].values.tolist()[0])

data2 = {'LTTR1':[lr_avg_lttr1,rf_avg_lttr1,knn_ag_lttr1 ], 
        'LTTR2':[lr_avg_lttr2, rf_avg_lttr2, knn_ag_lttr2],
        'COV':[lr_avg_cov,rf_avg_cov, knn_ag_cov],
        'MUSH': [lr_avg_mush,rf_avg_mush,knn_ag_mush]} 
table_2 = pd.DataFrame(data2, index =['LR', 'RF', 'KNN']) 
display(table_2)
table_2.to_latex() #get latex code
```
